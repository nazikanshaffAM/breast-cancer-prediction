```
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data<(q1-1.5*IQR))|(data>. (q3+1.5*IQR)))]

  return outliers

```



````python
#function to detect outliers in A variable based on the interquartile range
def find_outliers_IQR(data_frame):
    q1=data_frame.quantile(0.25)
    q3=data_frame.quantile(0.75)
    IQR=q3-q1
    outliers = data_frame[((data_frame<(q1-1.5*IQR))|(data_frame>(q3+1.5*IQR)))]
    return outliers
````
